```python
# Finding locations of null values in this DataFrame
import numpy as np
null_locations = np.where(df.isnull())

# Print the locations of null values
for row, col in zip(null_locations[0], null_locations[1]):
    print(f"Null value at row {row} and column {df.columns[col]}")
```

Duplicate Values
---
Duplicates in a DataFrame refers to a row that have identical values in all columns as another row in the same DataFrame.

Duplicate rows can occur due to errors in data entry, data merging, or data extraction.

To determine the number of duplicates and their locations:

```python
# Finding the number of duplicates
duplicates=df.duplicated()
print(duplicates)
number_duplicates = df.duplicated().sum()

# Print the number of duplicates
print(f"There are {number_duplicates} duplicate rows in the DataFrame.")

# Find the duplicate rows
duplicate_rows = df[df.duplicated()]

# Print the duplicate rows
print("Duplicate rows in the DataFrame:")
print(duplicate_rows)
```

```python
# Print the original DataFrame
print('Original DataFrame:\n',df)

# Remove null count values
# Remove any rows with missing values
df = df.dropna()

# Print the DataFrame without null count values
print('DataFrame without null count values:\n', df)
```

```python
#Print the original DataFrame
print('Original DataFrame:\n',df)


# Remove duplicate rows
df = df.drop_duplicates()

# Print the DataFrame without duplicates
print('DataFrame without duplicates:\n',df)
```

This analysis shows there are no duplicates.

### Showing Data Correlation Among The Columns


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#to calculate the correlation matrix
#to plot the heatmap
col = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol','quality']
cor= df[col].corr()
fig,ax = plt.subplots(figsize=(11, 6)) 
heatmap = sns.heatmap(cor, annot=True, cmap='viridis',ax=ax)
heatmap.set_xticklabels(heatmap.get_xticklabels())
```

### Histogram And Piecharts


```python
import seaborn as sns
histogram_plot1 = sns.histplot(data=df, x='fixed acidity')

histogram_plot1.set_title("Histogram plot for fixed acidity")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mean_1 = df['fixed acidity'].mean()


above_mean_count = df[df['fixed acidity'] > mean_1]['fixed acidity'].count()
below_mean_count = df[df['fixed acidity'] < mean_1]['fixed acidity'].count()


labels = ['Greater than Mean', 'Less than Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['green', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%' )
# plt.axis('equal')
plt.title('Fixed Acidity Greater than and Less than Mean')
plt.show()
```
